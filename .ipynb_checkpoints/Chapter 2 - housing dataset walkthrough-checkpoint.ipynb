````python
import os
import tarfile
from six.moves import urllib
````

````python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url = HOUSING_URL, housing_path = HOUSING_PATH):
    
    if not os.path.isdir(housing_path): # define path using os.path <- check if there's a path called housing_path
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, 'housing.tgz')
    urllib.request.urlretrieve(housing_url, tgz_path)
    
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path = housing_path)
    housing_tgz.close()

fetch_housing_data()
````

#### Stratified Sampling

This is sampled purely randomly through the dataset. However, suppose that `median_income` is believed to be a very important attribute to predict `median_house_value`, we may want to ensure the test/training set is representative of the various categories of incomes in the whole dataset. <br>
<br>
Hence we want to use stratified sampling here -- and before we do that, we need to regroup our `median_income` groups. <br>
In this example, the book suggests to create a income category by dividing `median_income` by 1.5 and re-categorize them into 1-5 groups.

````python
housing.median_income.hist()
````

````python
housing.median_income.describe()
````

````python
import numpy as np
housing['income_cat'] = np.ceil(housing.median_income/1.5)  #rounding numbers up, for new columns, using df['new_col']

housing.income_cat = housing.income_cat.apply(lambda x: x if x<5 else 5.0) # or you can use np.where here
housing.income_cat.value_counts()/len(housing)
````

````python
housing.income_cat.hist()
````

Now perform stratified sampling

````python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)

for train_index, test_index in split.split(housing, housing.income_cat):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)

def income_prop_results(data):
    return data.income_cat.value_counts()/len(data)

comparison = pd.DataFrame({'Overall':income_prop_results(housing),
                          'Stratified': income_prop_results(strat_test_set),
                          'Random':income_prop_results(test_set),
                          }).sort_index()

comparison['Rand. % Error'] = 100*comparison.Random/comparison.Overall - 100
comparison['Start. % Error'] = 100*comparison.Stratified/comparison.Overall -100

comparison
````

With that, we can now remove `income_cat` so the data would be back to its original form

````python
for set_ in (strat_train_set, strat_test_set):
    set_.drop(['income_cat'], axis = 1, inplace = True)
    
strat_train_set.head()
````

There are a few ways to handle missing values:
- Using Pandas `.dropna` , `.drop` or `.fillna` method (that col/series dont have to be numeric) & we have three choices typically:
    1. Get rid of the corresponding missing values within that column <br>
    *Sample Code:*
    ```python
    housing.dropna(subset = ['total_bedrooms']) 
    ```
    2. Get rid of the whole attribute <br>
    *Sample Code:*
    ```python
    housing.drop('total_bedrooms',axis =1)
    ```
    3. Set the values to some value (zero, mean, median, etc.) <br>
    *Sample Code:*
    ```python
    housing.total_bedrooms.fillna(median, inplace = True)
    ```

#### Handling Text and Categorical Attributes

We have `ocean_proximity` column leftover to convert into preferred format (categorical to numeric, i.e. one hot encoding) <br>

There are two Encoding transformer in Scikit-Learn, `LabelEncoder` and `OneHotEncoder`.

Note that `LabelEncoder` can only convert categorical variables into labels <br>
e.g.
```python
['dog','cat','turtle'] = [0,1,2]
```
This means the average of dog & turtle = cat, mathematically. Hence, for a lot of ML Algorithms, `OneHotEncoder` is more preferred. If we need those as multi-class labels for y, we can also use `LabelBinarizer`. This will give us something like: <br>

```python
['dog','cat','turtle'] = [[1 0 0], [0 1 0], [0 0 1]]
```

````python
# Example for LabelEncoder

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

housing_cat = housing.ocean_proximity
housing_cat_encoded = encoder.fit_transform(housing_cat)
print(housing_cat_encoded)

# check encoding classes
print(encoder.classes_)
````

Note that `fit_transform()` in expects a 2D array, but `housing_cat_encoded` is a 1D array, so we need to reshape it. Thus in currently work flow, we still need to use `LabelEncoder` to convert categorical variables to labels, reshape it and put it through `OneHotEncoder` <br>


````python
# Using OneHotEncoder

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse = False) #if set to true(default), this will return a sparse array
# with sparse array, we'll need to use .toarray() to convert it into a dense array

housing_cat_1hot = encoder.fit_transform(housing_cat_encoded.reshape(-1,1))
print(housing_cat_1hot)
print(encoder.active_features_)
````

Alternatively, we can use `LabelBinarizer`

````python
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer() # returns numpy dense array by default
housing_cat_1hot = encoder.fit_transform(housing_cat)
housing_cat_1hot
````

#### Custom Transformers
Here we want to add our own transformers for tasks such as custom cleanup operations or combining specific attributes. Meanwhile, this should work seamlessly with Scikit-learn functionalities (e.g. Pipelines). Hence we need to create a class and implement three methods: 
- `fit()` (returning self)
- `transform()`
- `fit_transform()` <= We can get this free by adding `TransformerMixin` as a base class
<br>
<br>
Also, if we add `BaseEstimator` as a base class, we'll get:
- `get_params()`
- `set_params()`
<br>
They will be useful for automatic hpyerparameter tuning.

````python
from sklearn.base import BaseEstimator, TransformerMixin

# column index
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix] # using indicies since input are ndarray
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)

housing_extra_attribs # this shoud be an array
````

Now create a data frame with extra attributes

````python
housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs, 
    columns = list(housing.columns) + ['rooms_per_household','population_per_household'] )
housing_extra_attribs.head()
````

#### Feature Scaling
Most ML algorithms don't perform well with input variables on different scale, hence feature scaling is necessary to standardize/normalize input variables. <br>
*Note: Scaling target values is generally not required* <br>
<br>
<br>
There are two common ways to get all attributes to have same scale:
- Min-max scaling (a.k.a. normalization): <br>
    `MinMaxScaler()` in Scikit-Learn <br>

- Standardization: <br>
    `StandardScaler()` in Scikit-Learn <br>

**It is important to fit scalers to training data only, not to the full dataset (including test set)**

#### Transformation Pipelines

With the dataset we have on hand, here is the pipeline we want to build out:
- Separate Data into different categories
    1. Numeric Columns
    2. Categorical Columns 
- Apply different treatment to each category
    1. Numeric: Scaling
    2. Categorical: One Hot Encoding
- Merge all treatments back together (as numpy array)

````python
# creating numeric pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imupter', Imputer(strategy = 'median')), # using median imupation for missing values
    ('attribs_adder', CombinedAttributesAdder()), # custom attribute adder to add bedroom/room by default
    ('std_scaler', StandardScaler()),  # scaling all numeric variables
])

housing_num_tr = num_pipeline.fit_transform(housing_num)

housing_num_tr
````

Earlier version of the book recommends using `DataFrameSelector` (created as custom class via `BaseEstimator` and `TransformerMixin`) and `FetureUnion`. It's now perferable to use `ColumnTransformer` that will be used in Scikit-Learn 0.20. For now, we can import it from `future_encoders.py`. After Scikit-Learn 0.20 is released, you can import it from `sklearn.compose` instead.

````python
from future_encoders import ColumnTransformer, OneHotEncoder # remember also to import OneHotEncoder here
````

````python
# the new way with scikit-learn 0.20
num_attribs = list(housing_num)  # grabbing numeric variable names
cat_attribs = ['ocean_proximity'] # grabbing categorical variable names

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs), # inserting numeric pipeline from above
    ('cat', OneHotEncoder(), cat_attribs),
])

housing_prepared = full_pipeline.fit_transform(housing)

print('housing_prepared shape:', housing_prepared.shape)
housing_prepared
````

#### Better Evaluation Using Cross-Validation

Using K-fold cross-validation via Scikit-Learn

````python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                        scoring = 'neg_mean_squared_error', cv = 10)

tree_rmse_scores = np.sqrt(-scores)
````

_Note: <br>
Scikit-Learn cross-validation features expect a utility function (greater = better) rather than a cost function (lower = better). Hence the scoring function is actually the opposite of the MSE (i.e. negative value), which is why the preceding code computes `-scores` before calculate square root._

Now comparing different models over training set using 10-fold cross validation:
- Decision Tree
- Linear Regression
- Random Forest

````python
def display_scores(scores):
    print('Scores: ', scores)
    print('Mean: ', scores.mean())
    print('Standard Deviation: ', scores.std())

#decision tree
display_scores(tree_rmse_scores)
````

#### Grid Search

The following `param_grid` tells the model to try out 3x4 combinations first, with `bootstrap = True` by default; <br>
then try out 2x3 combinations second with `bootstrap = False`

````python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators':[3,10,30],
    'max_features':[2,4,6,8],
    },
    {'bootstrap': [False],
     'n_estimators': [3,10],
     'max_features': [2,3,4]},
]

forest_reg = RandomForestRegressor(random_state = 42)

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                          scoring = 'neg_mean_squared_error', n_jobs = -1)

grid_search.fit(housing_prepared, housing_labels)
````

The best hyperparameter can be found via `.best_params_` method and the model can be accessed via `.best_estimator_` method

````python
grid_search.best_params_
````

*Note:<br>
Since 30 is the max value of `n_estimators` that was evaluated, we might want to evaluate higher values as well, as the score may continue to improve*

````python
grid_search.best_estimator_
````

To look at score of each hyperparameter combination:

````python
cv_res = grid_search.cv_results_

for mean_score, params in zip(cv_res['mean_test_score'], cv_res['params']):
    print(np.sqrt(-mean_score), params)
````

#### Randomized Search

#### Evaluate Your System on the Test Set

Now it's time to test our model over test set to see the performance of our model. <br>
<br>
*Note:<br>
Run `full_pipeline` to transform the data ( __call `transform()`, not `fit_transform()`__ )*

````python
final_model = grid_search.best_estimator_
X_test = strat_test_set.drop('median_house_value', axis = 1)
y_test = strat_test_set.median_house_value.copy()

X_test_prepared = full_pipeline.transform(X_test)
final_prediction = final_model.predict(X_test_prepared)

final_mse = mean_squared_error(y_test, final_prediction)
final_rmse = np.sqrt(final_mse)

final_rmse
````

We can compute a 95% confidence interval for the test RMSE:

````python
from scipy import stats

confidence = 0.95
squared_errors = (final_prediction - y_test)**2
mean = squared_errors.mean()
m = len(squared_errors)

np.sqrt(stats.t.interval(confidence, m-1,
                        loc = np.mean(squared_errors),
                        scale = stats.sem(squared_errors)))
````

````python
from scipy.stats import expon, reciprocal

# see https://docs.scipy.org/doc/scipy/reference/stats.html
# for `expon()` and `reciprocal()` documentation and more probability distribution functions.

# Note: gamma is ignored when kernel is "linear"

param_distribs = {
    'kernel':['linear','rbf'],
    'C':reciprocal(20,200000),
    'gamma':expon(scale = 1.0),
}

svm_reg = SVR()

svm_rnd_search = RandomizedSearchCV(svm_reg,param_distributions= param_distribs,
                                    n_iter = 50, cv = 5, scoring = 'neg_mean_squared_error',
                                    verbose = 2, n_jobs = -1, random_state = 42
                                    )

svm_rnd_search.fit(housing_prepared, housing_labels)
````

Typically Randomized search tends to find better hyperparameters than grid search within the same amount of time.<br>
<br>
Also, let's take a look at exponential distribution we used (for `'gamma'`), with `scale = 1.0`. Note that some samples are much larger or smaller than 1.0, but when you look at the log of the distribution, you can see that moster values are actually concentrated roughly in the range of exp(-2) to exp(+2), which is about 0.1 to 7.4.

````python
expon_distrib = expon(scale = 1.0)

samples = expon_distrib.rvs(10000, random_state = 42)

plt.figure(figsize = (10,4))
plt.subplot(121)
plt.title('Exponential Distribution (scale = 1.0)')
plt.hist(samples, bins = 50)
plt.subplot(122)
plt.title('Log of Exponential Distribution')
plt.hist(np.log(samples), bins = 50)
````

Similarly let's take a look at distribution we used for `C` ( a uniform distribution within a given range):

````python
reciprocal_graph = reciprocal(20,200000)
samples = reciprocal_graph.rvs(10000, random_state = 42)


plt.figure(figsize = (10,4))
plt.subplot(121)
plt.title('Reciprocal Distribution')
plt.hist(samples, bins = 50)
plt.subplot(122)
plt.title('Log of Reciprocal Distribution')
plt.hist(np.log(samples), bins = 50)
````
